# Analyzing local anomaly correlations...

```python
import pandas as pd
import tables as tb
import numpy as np
from IPython.display import Image
%load_ext autoreload
%autoreload 1
%aimport LIMTools
```

```python
h5f = tb.open_file('/home/chaos2/wperkins/data/pyLIM/LIM_data.h5', 'r')
corr = h5f.root.stats.corr.read()
ce = h5f.root.stats.ce.read()
lats = h5f.root.data.lats.read()
lons = h5f.root.data.lons.read()
lats, lons = np.meshgrid(lats, lons,  indexing='ij')
```

```python
#1 year forecasts
data = corr[1:2] #~1 year forecast
mean_corr = data.mean(axis=0)
title = 'Local Anomaly Correlation: 1 year forecast'
filename = 'lca_1yr.png'
LIMTools.plot_corrdata(lats, lons, mean_corr.reshape(lons.shape), title, filename)
Image(filename=filename)
```

```python
#Test on single forecast time
data = corr[2:3]#~2 year forecast
title = 'Local Anomaly Correlation: 2-year forecast'
filename = 'lca_2tst.png'
LIMTools.plot_corrdata(lats, lons, data.reshape(lons.shape), title, filename)
Image(filename=filename)
```

```python
#2-5 year forecasts
data = corr[2:6] #~2-5 year forecast
mean_corr = data.mean(axis=0)
title = 'Local Anomaly Correlation: 2-5 year forecast'
filename = 'lca_2_5yr.png'
LIMTools.plot_corrdata(lats, lons, mean_corr.reshape(lons.shape), title, filename)
Image(filename=filename)
```
